```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
file_path = '/Users/drmorsy/Downloads/Kaggle_Dataset/Osteoporosis Risk Prediction/osteoporosis.csv'
```

```python
class preprocesser: 
    #this class reades, gives some insights about your data file, handels missing values and makes categorical data encoding.
    #after instantiating an object you must first call read method to make other methods in the class valid for use.
    def __init__(self):
        self.file_path = input("Write your file path")
    def reader(self):
        #This function takes files in xlsx, csv or json formate and read it and return the file formate.
        file_formate = self.file_path.split('.')[-1]
        if file_formate in  ('xlsx', 'csv', 'json'):
            print('type of this data is: {}.'.format(file_formate))
        else:
            return 'please enter file in xlsx, csv or json formate only'
        try:
            if file_formate == 'xlsx':
                self.df = pd.read_excel(self.file_path)
            elif file_formate == 'csv':        
                self.df = pd.read_csv(self.file_path)
            elif file_formate == 'json':        
                self.df = pd.read_json(self.file_path)    
            return self.df
        except:
            return 'we ahve no such file here.'

    def file_describer(self):
        #this function return the output of describe function on object and bool or int and float in tuple.
        #you can print out on of two outputs by indexing on function return value.
        object_bool = self.df.describe(include=[object, bool])
        int_float = self.df.describe()
        print(object_bool, 3*'\n', int_float)
        
    def calculate_mean_frequent(self):
        #this function gives mean of numeric column values and most frequent value of categorical columns
        columns_names = self.df.columns
        mean_list = []
        column_name_list_no=[]
        most_frequent = []
        column_name_list_ob = []
        for i in columns_names:
            if self.df[i].dtypes != 'object':
                column_name_list_no.append(i)
                mean_list.append(np.mean(self.df[i]))
            elif self.df[i].dtypes == 'object':
                column_name_list_ob.append(i)
                most_frequent.append(self.df[i].value_counts().idxmax())
        mean_df = pd.DataFrame(list(zip(column_name_list_no, mean_list)),columns=['Column_Name', 'Mean'])        
        M_F_df = pd.DataFrame(list(zip(column_name_list_ob, most_frequent)),columns=['Column_Name', 'Most_Frequent_Object'])        
        print(M_F_df,3*'\n',mean_df)

    def mean_or_mode(self,column, mean=True):
        #this function gives us mean of numeric column or mode of categorical column.
        try:
            if mean:
                print(self.df[column].mean())
            else:
                print(self.df[column].mode())
        except:
            print('Mode function handel object values only please set mean to False.')


    def handel_nan(self):
        return self.df.fillna(value=0)

    def categorical_data_incoder(self):
        object_index = []
        for i in self.df.columns:
            if self.df[i].dtype == 'object':
                object_index.append(self.df.columns.get_loc(i))
        ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), object_index )],
                         remainder='passthrough')
        return pd.DataFrame(ct.fit_transform(self.df))
        
    
        

```

```python
df = preprocesser()
```

```python
df.reader()
#read the data and get the file formate.
```

```python
df.file_describer()
#gives output of describe method on each categrical and numerical columns.
```

```python
df.calculate_mean_frequent()
#makes two dataframes, one for categrical column calculating the mode and one for numerical column calculating the mean.
```

```python
df.mean_or_mode('Age')    
#takes numerical column name as argument calculating the mean but if you will input categorical column you must set mean to False.
print('\n')
df.mean_or_mode('Vitamin D Intake', mean=False)    
```

```python
df.handel_nan()
#handels missing values.
```

```python
df.categorical_data_incoder()
#Categorical Data Encoding
```
